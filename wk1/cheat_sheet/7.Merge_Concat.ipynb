```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df.head()
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
total = pd.concat([df1, df2])
total.head()
```

```python
total = pd.concat([df1, df2], axis = 1)
total.head()
```

```python
total = pd.concat([df1, df2], axis = 0)
total.head(10)
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
Ans = pd.concat([df3, df4], axis = 0, join = 'inner').fillna(0)
Ans
```

```python
Ans = pd.concat([df3, df4], axis = 0, join = 'outer').fillna(0)
Ans
```
