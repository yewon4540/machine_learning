1. 데이터로드

```python
import pandas as pd
import numpy as np 

train = pd.read_csv('https://raw.githubusercontent.com/inrap8206/Bigdata_Analyst_Certificate_Korean/main/data/titanic/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/inrap8206/Bigdata_Analyst_Certificate_Korean/main/data/titanic/test.csv')
sub = pd.read_csv('https://raw.githubusercontent.com/inrap8206/Bigdata_Analyst_Certificate_Korean/main/data/titanic/submission.csv')
```

```python
train.head()
```

```python
train.info()
# Age, Cabin의 값이 89개가 아니므로 null을 포함하고있다.
```

2. 데이터 처리(X,y)

```python
y_train = train['Survived'] #label
X_train = train.drop('Survived', axis=1)
X_test = test.copy()
X_data = pd.concat([X_train, X_test]) #전처리를 위한 데이터셋 합치기 (이 경우엔 train과 test가 따로 존재했기에...)
X_data.drop(['Name', 'PassengerId'], axis=1, inplace=True)
```

3. 결측치 제거

```python
X_data['Cabin'].value_counts()
```

```python
X_data['Embarked'].value_counts()
```

```python
X_data['Age'].fillna(X_data['Age'].mean(), inplace=True)
X_data['Fare'].fillna(X_data['Fare'].mean(), inplace=True)
X_data['Cabin'].fillna('C23 C25 C27', inplace=True) # 204개밖에 없어서 통상적으로는 지우는게 보통이다.
X_data['Embarked'].fillna('S', inplace=True) # 결측치에 가장 많은 'S' 를 넣음
```

4. 범주형 변수 생성(get_dummies)

```python
X_data['Sex'] = X_data['Sex'].map(lambda x : 1 if x == 'female' else 0)
em_dic = {'S':1, 'Q':2, 'C':1}
X_data['Embarked'] = X_data['Embarked'].map(em_dic)

#one-hot encoding
X_data = pd.get_dummies(X_data)
X_data
```

5. 데이터분리

```python
X_train = X_data[:len(X_train)]
X_test = X_data[len(X_train):]
```

6. 연속형 변수 처리 (스케일링)

```python
from sklearn.preprocessing import StandardScaler # StandardScaler : 평균과 표준편차로 나눠주는 것
sc = StandardScaler()
train = sc.fit_transform(X_train)
test = sc.transform(X_test)
```

7. 모델 빌드

```python
# 알고리즘
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

# 객체변환(따지지말고 그냥 일케 하는거임.)
model1 = GradientBoostingClassifier(random_state=2021)
model2 = RandomForestClassifier(random_state=2021)
```

8. 성능 측정(정확도 측정)

```python
# test의 predict값을 추출한 뒤, test의 y값과 비교하여 정확도 측정

model1.fit(X_train, y_train)
preds1 = model1.predict_proba(X_test) # predict_proba : 확률로 변환

model2.fit(X_train, y_train)
preds2 = model2.predict_proba(X_test)

preds = (preds1*0.6 + preds2*0.4)/2

# 교차검증
score = cross_val_score(model1, X_train, y_train, cv=3, scoring='roc_auc') #자동화를 전제로 만들어진 함수임.
print(score)
print("평균 :", score.mean())
```

9. 모델 저장 (제출)

```python
sub['Survived'] = preds[:,1]
sub.to_csv('submission.csv', index=False)
```
