# Part.4

Apply_Map

[목차로 이동](#Cheat_Sheet)

```python
import pandas as pd 
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
df.shape
```

```python
df.head()
```

```python
df['Income_Category'].value_counts()
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e',
}
```

```python
df['newIncome'] = df['Income_Category'].map(lambda x : dic[x])
# map lambda
```

```python
df['newIncome'] = df['Income_Category'].map(dic)
# map dic
```

```python
df['newIncome']
```

```python
def changecategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K' :
        return 'a'
    elif x =='$40K - $60K': 
        return 'b'
    elif x =='$60K - $80K'    :
        return 'c'
    elif x =='$80K - $120K'   :
        return 'd'
    elif x =='$120K +'        :
        return 'e'

df['newIncome'] = df['Income_Category'].apply(changecategory)

df['newIncome']
```

```python
df['AgeState'] = df['Customer_Age'].apply(lambda x : x//10 * 10)
df['AgeState'].value_counts().sort_index()
```

```python
df['newEduLevel'] = df['Education_Level'].map(lambda x : 1 if 'Graduate' in x else 0)
```

```python
df['Education_Level'].value_counts()
```

```python
df['newEduLevel'].value_counts()
```

```python
df['newLimit'] = df['Credit_Limit'].map(lambda x : 1 if x >= 4500 else 0)
df['newLimit'].value_counts()
```

```python
def change_credit(x):
    if x >= 4500:
        return 1
    else : 
        return 0

df['newLimit'] = df['Credit_Limit'].map(change_credit)
```

```python
df['newLimit'].value_counts()
```

```python
def Check(x):
    if x['Marital_Status'] == 'Married' and x['Card_Category'] == 'Platinum' :
        return 1
    else :
        return 0

df['newState'] = df.apply(Check, axis = 1)
```

```python
df['newState'].value_counts()
```

```python
def ChangeGender(x):
    if x == 'M':
        return 'male'
    else :
        return 'female'

df['Gender'] = df['Gender'].apply(ChangeGender)
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].apply(lambda x : 'male' if x == 'M' else 'female')
df['Gender'].value_counts()
```

```python
def fix_century(x):
    import datetime

    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
```

```python
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

```python
df.head()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]

    if hour =='24':
        hour = '00:00:00'
        FinalDate = pd.to_datetime(date + " " + hour) + datetime.timedelta(days=1)

    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date + " " + hour)

    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
```

```python
df.head()
```

```python
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df.head()
#day_name : 요일표시
```

```python
ans1 = df.groupby(['dayName','PM10등급'], as_index = False).size()
ans1.head()
```

```python
ans1.pivot(index = 'dayName', columns = 'PM10등급', values = 'size').fillna(0)
```

```python
Ans = df.groupby(df['(년-월-일:시)'].dt.hour).mean().iloc[[10,22],[0]]
Ans
```

Merge_Concat

[목차로 이동](#Cheat_Sheet)

```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df.head()
#index_col : index 열 설정
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
total = pd.concat([df1, df2])
total.head()
```

```python
total = pd.concat([df1, df2], axis=1)

total.head()
```

```python
total = pd.concat([df1, df2], axis=0)

total.head()
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
Ans = pd.concat([df3, df4], axis = 0, join = 'inner').fillna(0)
Ans
```

```python
Ans = pd.concat([df3, df4], axis = 0, join = 'outer').fillna(0)
Ans
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
Ans = pd.merge(df5,df6,on='Algeria',how='inner')
Ans
# on=x : x중심으로
```

```python
Ans = pd.merge(df5,df6,on='Algeria',how='outer')
Ans
```
